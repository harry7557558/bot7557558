Load the official Unicode confusable characters list.

```python
import requests

r = requests.get("http://www.unicode.org/Public/security/revision-03/confusablesSummary.txt")
print(r.status_code)
raw_content = r.content
```

Get a list of confusable characters for each letter.

```python
# Unicode official list
content = raw_content.decode('utf-8').strip().replace('\r', '').split('\n\n')

# Manually go through [U+0020,U+05FF]
# Only includes those that are consistent across fonts
manually = """
AÀÁÂÃÄÅĀĂĄǞǠǺȀȂȦȺΆΑАѦaàáâãäåāăąǟǡǻȁȃȧа
BƁƂɃΒВbƀƃƄƅɓЪЬѢѣҌҍ
CÇĆĈĊČƇȻϹϾЄСҀҪcçćĉċčƈȼɕϲсҁҫ
DÐĎĐƉƊdďđƌȡɖԁ
EÈÉÊËĒĔĖĘĚȄȆȨɆΈΕЀЁЕeèéêëēĕėęěȅȇȩɇеѐёҽҿ
FƑϜfƒ
GĜĞĠĢƓǤǴԌgĝğġģǥǵɠɡց
HĤĦΉΗНҢҤӇӉԨhĥħɦɧћիհ
IÌÍÎÏĨĪĬĮİƗȈȊΊΙΪІЇӀӏiìíîïĩīĭįıȉȋɨίιϊії
JĴɈЈjĵȷɉɟϳјյ
KĶƘΚϏЌКҚҜҞҠӃԞkķĸƙқҟԟ
LĹĻĽĿŁȽԼlĺļľŀłƖƚɫɬ
MΜϺМӍmɱӎ
NÑŃŅŇƝǸΝЍИЙҊӢnñńņňŉŋǹɲɳήηդղոր
OÒÓÔÕÖŌŎŐƠǪǬǾȌȎȪȬȮȰΌΘΟОѲѺӨՕoòóôõöōŏőơǫǭǿȍȏȫȭȯȱοσόϙоѳѻөծօ
PƤΡРҎpƿρϼрҏ
QɊԚqɋԛգզ
RŔŖŘƦȐȒɌЯrŕŗřȑȓɍ
S$ŚŜŞŠȘЅՏsśŝşšșȿѕ
TŢŤŦƬȚȾΤТҬԎtţťŧƫț
UÙÚÛÜŨŪŬŮŰŲƯƲǕǗǛȔȖɄԱՄՍuùúûüũūŭůűųưǖǘǜȕȗΰυϋύս
VѴѶvɣνѵѷ
WŴԜwŵωώѡѿԝ
XΧЖХҖҲӼӾxχжхҗҳӽӿ
YÝŶŸƳȲɎΎΥΫҮҰyýÿŷƴȳɏуўүұӯ
ZŹŻŽƵȤΖzźżžƶȥɀʐʑ
""".strip().split('\n')


letter_chars = {}

for letter in "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz":
    lowered = letter.lower()
    if lowered not in letter_chars:
        letter_chars[lowered] = []
    # Unicode official list
    for block in content:
        if not block.startswith("#\t"+letter):
            continue
        chars = block.split('\n')[0].split('\t')[1:]
        if len(chars[0]) > 1 and chars[0][1].isalpha():
            continue
        chars = [c for c in chars if len(c) == 1]
        letter_chars[lowered] += chars
    # manual work
    for line in manually:
        if letter in line:
            letter_chars[letter.lower()] += list(line)
    letter_chars[lowered] = list(set(letter_chars[lowered]))


print(letter_chars)
```

Remove characters than can be filtered by Python `unicodedata`.

```python
import unicodedata
import re


def clean(text):
    text = unicodedata.normalize('NFKD', text)
    text = re.sub(
        r"[\u0300-\u036F\u1AB0-\u1AFF\u1DC0-\u1DFF\u20D0-\u20FF\uFE20-\uFE2F\.\,·]",
        '', text)
    # if len(text) != 1:
    #     print(text)
    return text[0]


letter_sets = {}

for (letter, chars) in letter_chars.items():
    unique_chars = set({})
    for char in chars:
        char = clean(char)
        if not re.match(r"[A-Za-z\s]", char):
            unique_chars.add(char)
    letter_sets[letter] = ''.join(unique_chars)

for letter in "abcdefghijklmnopqrstuvwxyz":
    assert letter in letter_sets

print(str(letter_sets).replace(', ', ',\n').replace('{', '').replace('}', ''))
```
