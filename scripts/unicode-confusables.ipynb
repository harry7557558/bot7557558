Load the official Unicode confusable characters list.

```python
import requests

r = requests.get("http://www.unicode.org/Public/security/revision-03/confusablesSummary.txt")
print(r.status_code)
raw_content = r.content
```

Get a list of confusable characters for each letter.

Remove characters than can be filtered by Python `unicodedata`.

```python
import unicodedata
import re


def clean(text):
    text = unicodedata.normalize('NFKD', text)
    text = re.sub(
        r"[\u0300-\u036F\u1AB0-\u1AFF\u1DC0-\u1DFF\u20D0-\u20FF\uFE20-\uFE2F\.\,Â·]",
        '', text)
    # if len(text) != 1:
    #     print(text)
    return text[0]


letter_sets = {}

for (letter, chars) in letter_chars.items():
    unique_chars = set({})
    for char in chars:
        char = clean(char)
        if not re.match(r"[A-Za-z\s]", char):
            unique_chars.add(char)
    letter_sets[letter] = ''.join(unique_chars)

for letter in "abcdefghijklmnopqrstuvwxyz":
    assert letter in letter_sets

print(str(letter_sets).replace(', ', ',\n').replace('{', '').replace('}', ''))
```
